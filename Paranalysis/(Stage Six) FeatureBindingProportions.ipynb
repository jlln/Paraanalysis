## This script compared the proportions of clusters overlapping each feature type for the two experimental datasets and HurC (taken from a published ParClip dataset                                             

### (Lebedeva, 2011.Transcriptome-wide analysis of regulatory interactions of the RNA-binding protein HuR. Mol Cell. 2011 Aug 5;43(3):340-52. doi: 10.1016/j.molcel.2011.06.008)

```python
import pandas

directory='./MiSeqSFPQc.csv_MiSeqNONOc.csv_HuRc_conversioncTranscripts/'

hurc=pandas.read_csv(directory+'HuRc_conversioncfeature_cluster_table.csv')
SFPQ=pandas.read_csv(directory+'MiSeqSFPQc.csvfeature_cluster_table.csv')
NONO=pandas.read_csv(directory+'MiSeqNONOc.csvfeature_cluster_table.csv')

datasets=[hurc,SFPQ,NONO]
for x in datasets:
    del x['Unnamed: 0']
    x.iloc[:,1]=x.iloc[:,1].div(x.iloc[:,1].sum(axis=1), axis=0)*100
    sample=[x.columns[1]]*len(x.index)
    x['Sample']=sample
    x.columns=[u'Feature_Type', u'Percentage of Clusters in this Feature Type', u'Sample']
data=pandas.concat(datasets)
data=data.fillna(0)

    
    
```

```python
data
```

```python
Translation_Status=[]
Feature_Type=[]
def feature_classifier(row):
    feature=row['Feature_Type']
    if "ncRNA" in feature:
        Translation_Status.append("ncRNA")
    else:
        Translation_Status.append("mRNA")
    if "UTR" in feature:
        Feature_Type.append("UTR")
    elif "Exon" in feature:
        Feature_Type.append("Exon")
    elif "Intron" in feature:
        Feature_Type.append("Intron")
    elif "Splice" in feature:
        Feature_Type.append("Splice Junction")
    elif "Start" in feature:
        Feature_Type.append("Translation Start")
    
data.apply(feature_classifier,axis=1)
    
```

```python
data['Transcript_Type']=Translation_Status
data['Feature_Category']=Feature_Type
```

```python
data
```

```python
data=data.sort(['Feature_Category'])
```

```python
#Save a csv for later follow up analysis
data.to_csv(directory+"featurebindingproportions.csv")
```

### The subgroups were consolidated into their overall feature categories for plotting.

```python
feature_groups=data.groupby(['Sample','Transcript_Type','Feature_Category'])
```

```python
feature_groups.sum().reset_index().to_csv('featurebindingproportiontable.csv')
```

```python
feature_groups=feature_groups.sum().reset_index()
```
